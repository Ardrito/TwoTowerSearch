```python
from datasets import load_dataset
import random
from nltk import word_tokenize
import re
from tqdm import tqdm
import pickle

random.seed(42)
```

```python
marco_dataset = load_dataset("ms_marco", "v1.1")
```

```python
marco_dataset['validation'].shape
```

```python
text = {}

for split in marco_dataset:
    print (split)
    for sample in marco_dataset[split]:
        for words in (sample['passages']['passage_text']):
            words_ = re.sub(r'[^a-zA-Z-\s]', '', words)
            words_ = (words_.lower().split(" "))
            
            for word in words_:
                if word in text:
                    text[word] += 1
                else:
                    text[word] = 0
            
            
            
```

```python
tockenizer = {}
counter = 0
for word in text:
    if text[word] >= 10:
        counter += 1
        tockenizer[word] = counter
```
